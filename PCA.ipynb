```r
# import data
data <- read.csv('/Users/alessia/Desktop/economic_freedom/data.transformed.csv')
# only select numeric variables
num.data <- data[c(4:16)]
```

```r
# correlation matrix of transformed variables
library(corrplot)
corrplot.mixed(cor(num.data), tl.col = 'black', tl.cex = 0.75)
```

```r
# PCA analysis
library(rrcov)
data.pca <- PcaClassic(num.data, scale = F) # without scaling
summary(data.pca)
data.pca$loadings
corrplot(data.pca$loadings, is.corr = F, method = 'color', tl.col = 'black', tl.cex = 0.75, addgrid.col = 'grey')
corrplot(cor(num.data, data.pca$scores), tl.col = 'black', tl.cex = 0.75)
```

The first and second principal component capture together over 97% of the variance. Looking at the loading vectors, we notice that the first PC is mainly comprised of the variables Labor Freedom, whereas the second PC is mainly comprised of the variables Business Freedom and Overall Score.

```r
# PCA analysis
library(rrcov)
data.pca <- PcaClassic(num.data, scale = T) # with scaling
summary(data.pca)
data.pca$loadings
corrplot(data.pca$loadings, is.corr = F, method = 'color', tl.col = 'black', tl.cex = 0.75, addgrid.col = 'grey')
corrplot(cor(num.data, data.pca$scores), tl.col = 'black', tl.cex = 0.75)
```

If we scale the variables before performing PCA, we notice that the more variables contribute to the first and second principal components. In this case, two components explain around 68% of the variance. In order to choose an appropriate number of components, we can use a scree plot.

```r
# scree plot to choose the number of components
data.pca <- PcaClassic(num.data, scale = TRUE) # we need to scale the data -> correlation matrix
summary(data.pca)
screeplot(data.pca, type = 'lines', pch = 19, main = 'Scree plot')
```

According to the "elbow" in the scree plot, an appropriate number of components would be 3 or 4. We decide to select 4, in order to explain around 80% of the total variance.

```r
# select 4 PCs
data.pca.4 <- PcaClassic(num.data, scale = T, k = 4)
summary(data.pca.4)
```

```r
# scatterplots of scores
# create temporary dataframe
data.temp <- data
data.temp$Region <- as.numeric(as.factor(data$Region))
```

```r
# Region on pca
colors <- c('blue', 'red', 'green', 'orange', 'purple')
par(pty="s")
pairs(data.pca.4$scores,asp=1, col = colors[unclass(data.temp[,'Region'])], main = 'PCA scores grouped by Region')
par(pty="m")
summary(data.pca)
plot(data.pca.4$scores[,1:2], pch = 19, asp = 1, col = colors[unclass(data.temp[,'Region'])]) # zoom in on first 2 PCAs
```

By coloring the projected instances on the new lower-dimensional space by Region, we notice that the green group, which corresponds to Europe, appears to be more easily separable when considering the first two principal components. This is in accordance with the previous visualization.

```r
# biplot of components and scores based on the categorical variable Region
library(factoextra)
data.pca <- prcomp(num.data, scale = T)
fviz_pca_biplot(data.pca,
                label="var",
                habillage = data$Region,
               xlim=c(10, -10))
```

Looking at the biplot, we see that Europe is easier to separate from other Regions, as was also evident from the boxplots visualization by the categorical variable Region. Taking a look at variables, we notice that Tax Burden, Government Spending and Fiscal Health differ from other variables, contributing to the second principal component, rather than to the first one, as was also evident through the loading vectors. This distinction in the predictors was already evident in the initial correlation matrix, where Tax Burden, Fiscal Health and Government Spending did not appear to be correlated to any other variable.
